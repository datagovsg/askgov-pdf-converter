```python
import pandas as pd

df = pd.read_csv('datasets/askjaime3months.csv')
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df
```

```python
df = df[df['answer_taught'] != 'Catch all']
df
```

```python
import dialogflow_v2
from google.oauth2 import service_account

creds = service_account.Credentials.from_service_account_file('askjaime-creds.json')
```

```python
agents_client = dialogflow_v2.AgentsClient(credentials = creds)
agents_parent = agents_client.project_path(creds._project_id)

# intents_client = dialogflow_v2.IntentsClient(credentials = creds)
# intents_parent = intents_client.project_agent_path(creds.project_id)

# entity_types_client = dialogflow_v2.EntityTypesClient(credentials = creds)
# entity_types_parent = entity_types_client.project_agent_path(creds.project_id)

response = agents_client.get_agent(agents_parent)
response
```

```python
sessions_client = dialogflow_v2.SessionsClient(credentials = creds)
session = sessions_client.session_path(creds._project_id, '12345')
```

```python
def query_input(text):
    return {
        'text': {
            'text': text,
            'language_code': 'en',
        },
    }

response = sessions_client.detect_intent(session, query_input('parking boundary'))
response.query_result.fulfillment_text
```

```python
from ratelimit import limits, RateLimitException, sleep_and_retry
from backoff import on_exception, expo
from tqdm import tqdm

# @on_exception(expo, RateLimitException, max_tries=8)
@sleep_and_retry
@limits(calls=160, period=60)
def detect_intent(text):
    response = sessions_client.detect_intent(session, query_input(text))
    return response

dialogflow_answers = []

for row in tqdm(df.iterrows(), total=df.shape[0]):
    qn = row[1]['question_asked']
    result = detect_intent(qn)
    ans = result.query_result.fulfillment_text
    dialogflow_answers.append(result.query_result.fulfillment_text)

dialogflow_answers[:20]
```
